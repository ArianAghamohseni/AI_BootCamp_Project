```python
df_1 = df
```

```python
from sklearn.preprocessing import OneHotEncoder


categorical_features_1 = ['county', 'grspan', 'district']

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df_1[categorical_features_1]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features_1))

df_1 = df_1.drop(categorical_features_1, axis=1)
df_1 = pd.concat([df_1, encoded_df], axis=1)

df_1 = df_1.astype(int)
```

```python
df.drop(["district"], axis=1, inplace=True)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
اصلاحات مورد نیاز رو انجام داده و دوباره انکودینگ رو انجام بدید</font>
</p>

```python
from sklearn.preprocessing import OneHotEncoder


categorical_features = ['county', 'grspan']

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[categorical_features]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))

df = df.drop(categorical_features, axis=1)
df = pd.concat([df, encoded_df], axis=1)

df = df.astype(int)
```

```python
df
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
آماده کردن مدل اصلی</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
با توچه دیتایی که ساختید و فیچر هایی که ازش استخراج کردید مدل رو پیاده کنید</font>
</p>

```python
import statsmodels.api as sm


y = df['mathscr']
X = df.drop(['mathscr'], axis=1)
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
توضیف مدل</font>

<font face="vazir" size=3>
بر اساس جدولی که از متد model.summary() بدست اماده، کارایی مدل را تحلیل کنید

</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
میزان اهمیت هر متغیر برای مدل را تعیین کرده

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">

</p>

```python
print(anova_results)
```
